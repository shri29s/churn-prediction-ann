```python
import pandas as pd
import numpy as np
```

```python
from tensorflow.keras.models import load_model
```

```python
model = load_model("model.keras")
```

```python
model.summary()
```

```python
import pickle

col_trans = None
with open("col_trans.pkl", "rb") as file:
    col_trans = pickle.load(file)
```

```python
test_data = {
    "CreditScore": 600,
    "Geography": "France",
    "Gender": "Female",
    "Age": 19,
    "Tenure": 3,
    "Balance": 20000,
    "NumOfProducts": 3,
    "HasCrCard": 1,
    "IsActiveMember": 0,
    "EstimatedSalary": 40000
}
```

```python
pd.DataFrame(test_data, index=[0])
```

```python
test_data = col_trans.transform(pd.DataFrame(test_data, index=[0]))
```

```python
test_data.shape
```

```python
predictions = model.predict(test_data)
```

```python
predictions[0][0]
```

```python
if predictions[0][0] > 0.5:
    print("Customer is likely to churn")
else:
    print("Customer is not likely to churn")
```
